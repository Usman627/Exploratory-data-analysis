# Feature understanding

```python
df['Year_Introduced'].value_counts().head(10)
```

```python
#plotting top 10 years where most rides were introduced
ax = df['Year_Introduced'].value_counts().head(10)\
    .plot(kind = 'bar', title='Top 10 years Coasters were introduced in',figsize=(10,5))
ax.set_xlabel('Year introduced')
ax.set_ylabel('Count')
plt.show()
```

```python
#ploting the frequency of speed of all rides
ax = df['Speed_mph'].plot(kind='hist', bins=20, 
                          title='Frequency of Coster speed (mph)', figsize=(10,7))
ax.set_xlabel('Speed(mph)')
plt.show()
```

```python
#plotting top 10 manufacturers of these rides
ax = df['Manufacturer'].value_counts().head(10)\
    .plot(kind = 'barh', title='Top 10 Manufacturers',figsize=(10,6))
ax.set_xlabel('Count')
ax.set_ylabel('Manufacturer')
plt.show()
```

# Feature relationships

```python
#plotting scatter plot to see relation between speed vs height
df.plot(kind='scatter', x='Speed_mph', y='Height_ft', 
        title='Coaster speed vs Height', figsize=(10,6))
plt.show()
```

```python
#plotting seaborn scatter plot to see relation between speed vs height
ax = sns.scatterplot(x='Speed_mph',
                y='Height_ft',
                hue='Year_Introduced',
                data=df)
ax.set_title('Coaster Speed vs. Height')
sns.set(rc={'figure.figsize':(10,5)})
plt.show()
```

```python
#pair plot to see further relations between other factors as well
sns.pairplot(df, 
             vars=['Year_Introduced','Speed_mph','Height_ft',
                   'Inversions','Gforce'], hue='Type_Main')
plt.show()
```

```python
#correlation and heatmap
df_corr = df[['Year_Introduced','Speed_mph','Height_ft','Inversions','Gforce']]\
            .dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, cmap='Reds', annot=True)
plt.show()
```

# Queries

```python
#Locations with highest avg speed and also have 10 rides minimum

dfq = df.query('Location != "Other"')\
    .groupby('Location')['Speed_mph']\
    .agg(['mean','count'])\
    .query('count >= 10')\
    .sort_values('mean')\
    .plot(kind='barh',title='Avg speed by location',figsize=(10,6))
dfq.set_xlabel('Avg speed')
plt.show()
```

```python
#manufacturer with fastest rides, min 25 rides
dfq = df.groupby('Manufacturer')['Speed_mph']\
.agg(['mean', 'count'])\
.query('count>=25').sort_values('mean')['mean']\
.plot(kind='barh',title='Top manufacturer that produces fasted rides',figsize=(10,6))
dfq.set_xlabel('Avg speed')
plt.show()
```
